```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial
```

# Exercise:  Atmospheric carbon dioxide rise

```python
# Assuming 'years' is the array of years and 'co2_levels' is the array of CO2 levels
# For example:
# years = np.array([1980, 1981, 1982, ..., Last Year])
# co2_levels = np.array([338.7, 340.1, 341.4, ..., Last CO2 Level])

# We will generate some example data for demonstration; replace this with your actual data
years = np.arange(1980, 2021)
co2_levels = 0.1 * (years - 1980) ** 2 + 0.5 * (years - 1980) + 338.7 + np.random.normal(0, 1, len(years))

# Fit the quadratic curve (second-degree polynomial)
coefs = np.polyfit(years, co2_levels, 2)
p = np.poly1d(coefs)

# Calculate the derivative of the quadratic curve
derivative = np.polyder(p)

# Plot the data and the fitted curve
plt.scatter(years, co2_levels, label='Data', color='blue')
plt.plot(years, p(years), label='Quadratic Fit', color='red')
plt.title('Quadratic fit to CO2 levels')
plt.xlabel('Year')
plt.ylabel('CO2 Levels')
plt.legend()
plt.show()

# Estimate the rate of change at the last year
last_year_rate_of_change = derivative(years[-1])

# Provide a "yes or no" answer to whether the rate of increase of CO2 is increasing
answer = "Yes" if coefs[0] > 0 else "No"

print(f"The coefficients of the quadratic fit are: {coefs}")
print(f"The rate of change in the last year is approximately: {last_year_rate_of_change:.2f} ppm/year")
print(f"Is the rate of increase of CO2 levels increasing? {answer}")
```
