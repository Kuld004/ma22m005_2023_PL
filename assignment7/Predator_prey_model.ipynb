```python
import numpy as np
import matplotlib.pyplot as plt
from random import uniform
```

```python
# Constants
alpha = 1.1
beta = 0.4
delta = 0.1
gamma = 0.1
```

#Task:1

```python
# Initial conditions
r0 = 5.0
x0 = 2.0
```

```python
# Time parameters
t0 = 0
t_max = 70
dt = 0.01  # Time step
```

```python
# Arrays to store the results
t_values = [t0]
r_values = [r0]
x_values = [x0]
```

```python

# Euler method for numerical integration
while t0 < t_max:
    # Calculate the derivatives
    drdt = alpha * r0 - beta * r0 * x0
    dxdt = delta * r0 * x0 - gamma * x0

    # Update r and x using Euler's method
    r0 += drdt * dt
    x0 += dxdt * dt

    # Update time
    t0 += dt

    # Append values to the arrays
    t_values.append(t0)
    r_values.append(r0)
    x_values.append(x0)

```

```python
# Plot the solutions
plt.figure(figsize=(10, 6))
plt.plot(t_values, r_values, label='Rabbits (r)')
plt.plot(t_values, x_values, label='Foxes (x)')
plt.xlabel('Time (t)')
plt.ylabel('Population')
plt.title('Predator-Prey Model: Rabbits vs Foxes')
plt.legend()
plt.grid(True)
plt.show()
```

#Task:3

```python


# Define the equations for dr/dt and dx/dt
def drdt(r, x):
    return alpha * r - beta * r * x

def dxdt(r, x):
    return delta * r * x - gamma * x

# Function to find equilibria
def find_equilibria():
    # Solve for equilibrium points (r_eq, x_eq)
    r_eq = (gamma / delta)
    x_eq = (alpha / beta)
    return r_eq, x_eq

# Time parameters
t0 = 0
t_max = 70
dt = 0.01  # Time step

# Number of solutions to plot
num_solutions = 10

# Create a figure for the phase portrait
plt.figure(figsize=(10, 6))
plt.xlabel('Foxes (x)')
plt.ylabel('Rabbits (r)')
plt.title('Predator-Prey Model Phase Portrait')

# Find and plot the equilibria
r_eq, x_eq = find_equilibria()
plt.plot(x_eq, r_eq, 'ro', label='Equilibrium Points')

# Generate and plot 10 solutions with random initial values
for _ in range(num_solutions):
    r0 = uniform(1, 9)
    x0 = uniform(1, 9)

    # Arrays to store the results
    t_values = [t0]
    r_values = [r0]
    x_values = [x0]

    # Euler method for numerical integration
    while t0 < t_max:
        # Calculate the derivatives
        dr = drdt(r0, x0)
        dx = dxdt(r0, x0)

        # Update r and x using Euler's method
        r0 += dr * dt
        x0 += dx * dt

        # Update time
        t0 += dt

        # Append values to the arrays
        t_values.append(t0)
        r_values.append(r0)
        x_values.append(x0)

    # Plot the trajectory in the phase plane
    plt.plot(x_values, r_values, label=f'Initial: r(0)={r_values[0]:.2f}, x(0)={x_values[0]:.2f}')

# Add a legend and grid
plt.legend()
plt.grid(True)

# Show the phase portrait
plt.show()
```
