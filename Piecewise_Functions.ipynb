```python
import numpy as np
import matplotlib.pyplot as plt
```

# Step 1: Generate uniformly spaced points in the interval [-5, 5]

```python

num_points = 1000001
x = np.linspace(-5, 5, num_points)
```

# Step 2: Compute the function values based on the conditions

```python

# Using numpy's vectorized operations to efficiently compute the values
condition = (x < 0)
y = np.where(condition, x * np.sin(x**2), np.cos(x))
```

# Step 3: Time the computation (optional)

```python

import time
start_time = time.time()
```

# Step 4: Plot the function

```python

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x)', color='b')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Piecewise Function')
plt.grid(True)
plt.legend()
plt.show()
```

# Optional: Measure the execution time

```python

end_time = time.time()
execution_time = end_time - start_time
print(f"Execution time: {execution_time:.4f} seconds")
```
