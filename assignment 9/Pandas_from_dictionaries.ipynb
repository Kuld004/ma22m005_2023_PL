#3. Convert d1, d2 (together) to a pandas object dd.

```python

d1 = {'a': 1, 'b': 2}
d2 = {'b': 3, 'c': 4}

# Convert d1 and d2 into Pandas DataFrames
df1 = pd.DataFrame(list(d1.items()), columns=['Key', 'Value'])
df2 = pd.DataFrame(list(d2.items()), columns=['Key', 'Value'])

# Concatenate df1 and df2 into a single DataFrame dd
dd = pd.concat([df1, df2], ignore_index=True)

print(dd)
```
