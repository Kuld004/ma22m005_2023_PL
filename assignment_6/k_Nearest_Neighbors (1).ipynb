```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
```

# Task1

```python


def find_k_nearest_neighbors(points, k):
    """
    Find the k nearest neighbors for each point in the set of points.

    Args:
    points (numpy.ndarray): An array of shape (n, 2) representing the set of points in the plane.
    k (int): The number of nearest neighbors to find for each point.

    Returns:
    numpy.ndarray: An array of shape (n, k, 2) containing the k nearest neighbors for each point.
    """
    # Calculate pairwise Euclidean distances
    distances = cdist(points, points)

    # Sort distances for each point and select the k nearest neighbors
    sorted_indices = np.argsort(distances, axis=1)
    k_nearest_indices = sorted_indices[:, 1:k+1]  # Exclude the point itself

    # Extract the k nearest neighbors for each point
    k_nearest_neighbors = np.take(points, k_nearest_indices, axis=0)

    return k_nearest_neighbors


```

```python
# Define the points
P = np.array([[0, 0], [0.2, 0.22], [0.1, -0.1], [1, 1], [1.1, 0.9], [0.8, 0.9], [1.1, 0.63],
              [0.58, -0.1], [0.63, 0.1], [0.67, -0.3], [0.8, -0.23], [0.8, 0.6]])

# Specify the value of k
k = 3

# Find the k nearest neighbors for each point
k_neighbors = find_k_nearest_neighbors(P, k)

# Print the k nearest neighbors for each point
for i, neighbors in enumerate(k_neighbors):
    print(f"Point {i+1}: {neighbors}")
```

#Task2

```python


# Plot the points and arrows to k-nearest neighbors
plt.figure(figsize=(8, 6))
plt.scatter(P[:, 0], P[:, 1], marker='o', label='Points', color='blue')

for i, neighbors in enumerate(k_neighbors):
    for neighbor in neighbors:
        plt.arrow(P[i, 0], P[i, 1], neighbor[0] - P[i, 0], neighbor[1] - P[i, 1],
                  head_width=0.03, head_length=0.05, fc='red', ec='red')

plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'K-Nearest Neighbors with k={k}')
plt.grid(True)
plt.legend(['Points', f'{k}-Nearest Neighbors'])
plt.show()
```
