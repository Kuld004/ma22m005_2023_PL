```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Generate values for x from 0 to 10 with small increments
x = np.linspace(0, 10, 100)  # You can adjust the number of points (100) for smoother or coarser plotting.

# Calculate the corresponding y values using the sine function
y = np.sin(x)
```

```python


# Create a plot and add a label
plt.plot(x, y, label='sin(x)')  # Added label='sin(x)'

# Add labels and a legend
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of sin(x) for x in [0, 10]')
plt.legend()

# Show the plot
plt.show()
```
